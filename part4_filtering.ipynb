Operators for filtering:
- **&** - AND
- **|** - OR
- **~** - NOT

```python
filt = (attendance_df['Released'] == 0) & (attendance_df['School DBN'] == '01M015')
attendance_df.loc[~filt]
```

```python
sample_dbns = ['01M015', '01M019', '01M034']
dbn_filt = attendance_df['School DBN'].isin(sample_dbns)
attendance_df.loc[dbn_filt, 'School DBN'].value_counts()
```

```python
dbn_filt = attendance_df['School DBN'].str.contains('01M', na=False)
attendance_df.loc[dbn_filt, 'School DBN'].value_counts()
```
