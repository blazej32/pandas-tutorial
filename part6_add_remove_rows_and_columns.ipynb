```python
import pandas as pd
attendance_df = pd.read_csv('student_attendance.csv')
attendance_df.columns = ['school id', 'date', 'enrolled', 'absent', 'present', 'released']
attendance_df.columns = attendance_df.columns.str.replace(' ', '_')
attendance_df.columns = attendance_df.columns.str.upper()
attendance_df.head(10)
```

#### Join Columns

```python
attendance_df['SCHOOL_ID, DATE'] = attendance_df['SCHOOL_ID'] + ', ' + attendance_df['DATE'].apply(str)
attendance_df['ABSENT/RELEASED'] = attendance_df['ABSENT'] + attendance_df['RELEASED']
attendance_df
```

```python
attendance_df.drop(columns=['SCHOOL_ID, DATE'])
```

```python
attendance_df.drop(columns=['SCHOOL_ID, DATE', 'ABSENT/RELEASED'], inplace=True)
```
