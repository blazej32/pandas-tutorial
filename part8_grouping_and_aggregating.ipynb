```python
import pandas as pd
attendance_df = pd.read_csv('student_attendance.csv')
attendance_df.columns = ['school id', 'date', 'enrolled', 'absent', 'present', 'released']
attendance_df.columns = attendance_df.columns.str.replace(' ', '_')
attendance_df.columns = attendance_df.columns.str.upper()
attendance_df['%PRESENT'] = (attendance_df['PRESENT'] / attendance_df['ENROLLED'] * 100).round(2)
attendance_df.head(10)
```

```python
attendance_df['ABSENT'].mean()
```

```python
attendance_df.drop('SCHOOL_ID', axis=1).median()
```

```python
attendance_df.describe()
```

```python
attendance_df['SCHOOL_ID'].value_counts(normalize=True)
```

Group the dataframe values by SCHOOL_ID to perform further analysis

```python
groupby_school = attendance_df.groupby(['SCHOOL_ID'])
```

Let's check which school has the highest mean of presence percentage

```python
groupby_school['%PRESENT'].mean().sort_values(ascending=False).head(15)
```

Let's create some sort of summary table with presence % grouped by school_ids and selected aggregate functions

```python
groupby_school['%PRESENT'].agg(['count', 'median', 'mean', 'min', 'max']).sort_values(by='median', ascending=False).head(20)
```
