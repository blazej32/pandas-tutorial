```python
import pandas as pd
attendance_df = pd.read_csv('student_attendance.csv')
attendance_df.columns = ['school id', 'date', 'enrolled', 'absent', 'present', 'released']
attendance_df.columns = attendance_df.columns.str.replace(' ', '_')
attendance_df.columns = attendance_df.columns.str.upper()
attendance_df
```

#### Removing Missing Values

Remove missing values using df.dropna() method, axis=0 for rows, axis=1 for columns

```python
attendance_df.dropna(axis=0, how='any', subset=['SCHOOL_ID', 'DATE'] ,inplace=True)
attendance_df
```

```python
attendance_df.isna()
```

```python
attendance_df.fillna('MISSING')
```

You can also create a na_vals list at the beginning (e.g. na_vals = ['NA', 'Missing']) and then add pass it as an argument to read_csv function (..., na_values=na_vals)
